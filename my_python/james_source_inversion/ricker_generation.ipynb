## ricker generatimon - with delay 



```python
import numpy as np
import matplotlib.pyplot as plt
 
NtNew = 5800
dtNew = 4e-8 
    
def ricker(f, length=(NtNew*dtNew), dt=dtNew):
    t = np.arange(-length/2, (length-dt)/2, dt)
    y = - (1.0 - 2.0*(np.pi**2)*(f**2)*(t**2)) * np.exp(-(np.pi**2)*(f**2)*(t**2))
    return t, y
 
f = 500000 # A low wavelength of 25 Hz
t, w = ricker(f)


#time_delay = 500
#azeros = np.zeros(np.abs(time_delay))
#w_padded = np.append(azeros, w[t_star_show:t_end_show-time_delay], axis=0)


```

```python
t_new = np.arange(dtNew,5800*dtNew+dtNew,dtNew)
```

```python

one_signal_plot(t_new,w)
```

```python
## apply delay 
time_delay = 2000 
a = np.zeros(np.abs(time_delay))
#stf_inverted_trace_filtered_shifted = np.append(a, stf_inverted_trace_filtered[:time_step_comp], axis=0)
w_less_delay = np.append(w[time_delay:NtNew],a, axis=0)
```
